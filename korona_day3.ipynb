```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
plt.rcParams['figure.figsize'] = [15, 5]
from IPython import display
from ipywidgets import interact, widgets
```

```
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv',error_bad_lines=False)
```

```
df.head(10)
```

```
def top(date, top_n):
  return (df[['Province/State', 'Country/Region', date]].sort_values(by=date, ascending=True)).tail(top_n)
```

```
top('3/15/20', 5)
```

```
df.loc[df['Country/Region']=='China', :].sort_values(by='3/15/20', ascending=True).tail(10).head(9).style.background_gradient(cmap='RdBu_r')
```

```
def no_virus_until_given_date_inclusive(date):
  return df.loc[df[date]==0].index
```

```
no_virus_until_given_date_inclusive('3/15/20')
```

```
formated_gdf = df.groupby(['Country/Region']).max()
```

```
formated_gdf.sample(5)
```

```
formated_gdf = formated_gdf.reset_index()
```

```
formated_gdf.sample(10)
```

```
formated_gdf['size'] = formated_gdf['3/15/20'].pow(0.3) # normalizacja kek
```

```
formated_gdf.groupby
```

```
fig = px.scatter_geo(formated_gdf, locations="Country/Region", locationmode='country names', color="3/15/20", size="size", hover_name="Country/Region", range_color= [0, max(formated_gdf['3/15/20'])+2], projection="natural earth", \
                     title="Population of known cases of COVID-19 in the world")
fig.show()
```
